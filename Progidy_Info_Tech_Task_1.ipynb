## Create a bar chart or histogram to visualize the distribution of a categorical or continous variable, such as the distribution of ages or genders in a population

### Importing the necessary python libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

### Load and read the dataset

```python
# load the dataset etting header as 2 and using the column Indicator name as indexed column
df = pd.read_csv('/content/Population.csv', header=2, index_col="Indicator Name")
df
```

### Filter the data to get the least 5 countries by population in 2023

```python
# Use the sort values method to sort the dataset by 2023
# use head() to get the first 5 rows
small_countries = df.sort_values(by='2023', ascending=True).head()
small_countries[["Country Name","2023"]]
```

### Plot a bar chart to compare the population of the least 5 countries

```python
# Set the figure size
plt.figure(figsize=(8,5))

# plot a bar chart
plt.bar(small_countries["Country Name"],small_countries["2023"])

# add a title an xand y label
plt.title("Bar chart comparing the leat 5 countries by population in 2023")
plt.xlabel('Country Name')
plt.ylabel('Population')

# rotate the xticks
plt.xticks(rotation=45)

# save the figure
plt.savefig('Least_5_countries.png', dpi=300)
# show the plot
plt.show()
```

### Using a bar chart to compare the population of the least 5 countries in 2022 and 2023

```python
# creating a figure and axes
plt.figure(figsize=(10,4))
# create a width variable
width = 0.4

# set the x-axis positions
x = np.arange(len(small_countries['Country Name']))

# plot the multiple bar charts
plt.bar(x + width/2, small_countries["2023"], color='blue', label='2023', width=width)
plt.bar(x - width/2, small_countries["2018"], color='green',label='2018', width=width)

# set the title and the x and y labels
plt.title('A bar chart comparing of the population of the least 5 countries in 2023 to their population 5 years ago')
plt.xlabel('Country Name')
plt.ylabel('Population')

# set xticks
plt.xticks(x,small_countries['Country Name'], rotation =10)

# add a legend
plt.legend()
plt.show()
```

From the graph above the least 5 countries have not grown in population from their population 5 years ago.
